```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
import os

# --- Figure template (reuse your template) ---
fig_template = go.layout.Template()
fig_template.layout = {
    'template': 'simple_white+presentation',
    'autosize': False,
    'width': 800,
    'height': 600,
    'xaxis': {'ticks':'inside','mirror':'ticks','linewidth':2.5,'tickwidth':2.5,'ticklen':4,'showline':True,'showgrid':False},
    'yaxis': {'ticks':'inside','mirror':'ticks','linewidth':2.5,'tickwidth':2.5,'ticklen':4,'showline':True,'showgrid':False},
    'font': {'family': 'mathjax','size': 30}
}

# ============================================================
# --- Files to compare ---
# ============================================================
file1 = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\2nd Oct NaBiF4\Percentage Modulation\1064nm_2.75mA\1064nm_2.75mW_20Hz_975nm_100mW.txt"
file2 = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\3rd oct NaBiF4\same polarization\good Datas\1064nm_3.85mA_20Hz_975nm_300mW_recheck powers_500k.txt"


labels = ["File 1", "File 2"]
colors = ["royalblue", "crimson"]
markers = ["circle", "square"]

# --- Groups (use your existing groups)
groups = [(21,24),(46,49),(71,74),(96,99)]
t1, t2 = groups[0]  # group 1 window

# ============================================================
# --- Load data and align group 1 ---
# ============================================================
data_list = []

for fpath in [file1, file2]:
    arr = np.loadtxt(fpath, skiprows=1)
    if arr.ndim == 1 and arr.size >= 2:
        arr = arr.reshape(1, -1)
    time_ms = arr[:,0] / 1e9
    intensity = arr[:,1]
    data_list.append((time_ms, intensity))

# compute group 1 mean
mean_g1_file1 = np.nanmean(data_list[0][1][(data_list[0][0] >= t1) & (data_list[0][0] <= t2)])
mean_g1_file2 = np.nanmean(data_list[1][1][(data_list[1][0] >= t1) & (data_list[1][0] <= t2)])

# offset to align file2 to file1
offset = mean_g1_file1 - mean_g1_file2
time2, intensity2 = data_list[1]
intensity2_aligned = intensity2 + offset

# ============================================================
# --- Plotting ---
# ============================================================
fig = go.Figure()

# File 1
fig.add_trace(go.Scatter(
    x=data_list[0][0],
    y=data_list[0][1],
    mode='lines+markers',
    line=dict(color=colors[0], width=1.5),
    marker=dict(symbol=markers[0], size=6),
    name=labels[0],
    showlegend= False
))

# File 2 (aligned)
fig.add_trace(go.Scatter(
    x=time2,
    y=intensity2_aligned,
    mode='lines+markers',
    line=dict(color=colors[1], width=1.5),
    marker=dict(symbol=markers[1], size=6),
    name=labels[1],
    showlegend= False
))

# ============================================================
# --- Layout ---
# ============================================================
fig.update_layout(
    template=fig_template,
    xaxis_title="Time (ms)",
    # yaxis_title="Intensity (a.u.)",
    width=800,
    height=650
)
fig.update_xaxes( zeroline=False)


# Save figure
output_fig_folder = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\figures in paper"
os.makedirs(output_fig_folder, exist_ok=True)
out_path = os.path.join(output_fig_folder, "polarization_comparison_faceOn.png")
pio.write_image(fig, out_path, scale=4)
print("Saved figure:", out_path)

fig.show()
```
