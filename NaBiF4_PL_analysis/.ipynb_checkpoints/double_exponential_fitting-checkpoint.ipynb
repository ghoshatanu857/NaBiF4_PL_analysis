```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import plotly.graph_objects as go

# ============================================================
# File path
# ============================================================
file_path = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\2nd Oct NaBiF4\Percentage Modulation\1064nm_3.86mA\1064nm_3.86mW_20Hz_975nm_80mW.txt"

# ============================================================
# Load data (TAB-SEPARATED)
# ============================================================
df = pd.read_csv(file_path, sep="\t")

# Remove double quotes from column names
df.columns = [c.replace('"', '').strip() for c in df.columns]

print("Columns:", df.columns.tolist())

# Extract columns
time_ps = df["Time (ps)"].values
counts = df["Counts per bin"].values

# Convert ps  ms
time_ms = time_ps * 1e-9

# ============================================================
# Select fitting window (24.5 ms to 49.5 ms)
# ============================================================
t_min, t_max = 24.5, 49.5
mask = (time_ms >= t_min) & (time_ms <= t_max)
t_fit = time_ms[mask]
y_fit = counts[mask]

# ============================================================
# Double exponential function
# ============================================================
def double_exp(t, A1, tau1, A2, tau2, C):
    return A1 * np.exp(-t / tau1) + A2 * np.exp(-t / tau2) + C

# Initial guesses
p0 = [y_fit.max(), 0.09, y_fit.max()/5, 3.0, y_fit.min()]

# Fit
params, cov = curve_fit(double_exp, t_fit, y_fit, p0=p0, maxfev=20000)
A1, tau1, A2, tau2, C = params

print("\n=== Fitted Parameters ===")
print(f"A1   = {A1:.4e}")
print(f"tau1 = {tau1:.4e} ms")
print(f"A2   = {A2:.4e}")
print(f"tau2 = {tau2:.4e} ms")
print(f"C    = {C:.4e}")

# Smooth fitted curve (only inside fitting window)
t_smooth = np.linspace(t_min, t_max, 2000)
y_smooth = double_exp(t_smooth, *params)

# ============================================================
# Plot FULL RAW DATA + FITTED WINDOW
# ============================================================
fig = go.Figure()

# -------- Full raw data curve --------
fig.add_trace(go.Scatter(
    x=time_ms, y=counts,
    mode="markers",
    name="Raw Data (Full Range)",
    marker=dict(size=3, opacity=0.6)
))

# -------- Fitted curve (only inside window) --------
fig.add_trace(go.Scatter(
    x=t_smooth, y=y_smooth,
    mode="lines",
    name="Double Exp Fit (24.549.5 ms)",
    line=dict(width=3)
))

# -------- Highlight fitting window --------
fig.add_vrect(
    x0=t_min, x1=t_max,
    fillcolor="lightgrey", opacity=0.3,
    line_width=0,
    annotation_text="Fit Window",
    annotation_position="top left"
)

# -------- Plot layout --------
fig.update_layout(
    title="Full Raw Data with Double-Exponential Fit (24.549.5 ms)",
    xaxis_title="Time (ms)",
    yaxis_title="Counts per bin",
    template="simple_white",
    width=900,
    height=600
)

fig.show()
```
