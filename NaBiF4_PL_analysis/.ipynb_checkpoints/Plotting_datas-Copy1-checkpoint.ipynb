## Temperature Sensing

```python
import os
import numpy as np
import plotly.graph_objects as go

# --- Custom Plotly Template with axis label & tick label font sizes ---
fig_template = go.layout.Template()
fig_template.layout = {
    'template': 'simple_white+presentation',
    'autosize': False,
    'width': 800,
    'height': 600,
    'xaxis': {
        'ticks': 'inside',
        'mirror': 'ticks',
        'linewidth': 2.5,
        'tickwidth': 2.5,
        'ticklen': 4,
        'showline': True,
        'showgrid': False,
        'zerolinecolor': 'white',
        'titlefont': dict(size=34),
        'tickfont': dict(size=30),
    },
    'yaxis': {
        'ticks': 'inside',
        'mirror': 'ticks',
        'linewidth': 2.5,
        'tickwidth': 2.5,
        'ticklen': 4,
        'showline': True,
        'showgrid': False,
        'zerolinecolor': 'white',
        'titlefont': dict(size=34),
        'tickfont': dict(size=30),
    },
    'font': {'family': 'mathjax', 'size': 30}
}

# --- Path to folder ---
folder_path = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\2nd Oct NaBiF4\temperature sensing"

# --- Temperature list to include ---
temperatures = [25, 31, 35, 40, 45, 50, 55, 60]

# --- Temperatures that use the '_2.txt' version ---
use_2_suffix = {35, 45, 50, 55}

# --- Initialize figure ---
fig = go.Figure()
curves = []

# --- Load all data ---
for temp in temperatures:
    if temp in use_2_suffix:
        filename = f"1064nm_3.84mW_20Hz_975nm_220mW_{temp}degree_2.txt"
    else:
        filename = f"1064nm_3.84mW_20Hz_975nm_220mW_{temp}degree.txt"
    
    file_path = os.path.join(folder_path, filename)

    if not os.path.exists(file_path):
        print(f"⚠️ File not found: {filename}")
        continue

    # Load data (skip header, tab-separated)
    data = np.loadtxt(file_path, skiprows=1, delimiter='\t')
    time_ps = data[:, 0]
    counts = data[:, 1]

    curves.append((temp, time_ps, counts))

# --- Compute normalization factor based on average in 0–25 ms ---
t_low, t_high = 0, 25e9  # 25 ms = 25e9 ps
mean_values = []

for temp, time_ps, counts in curves:
    mask = (time_ps >= t_low) & (time_ps <= t_high)
    mean_in_range = np.mean(counts[mask]) if np.any(mask) else 1
    mean_values.append(mean_in_range)

# Reference mean = first curve’s mean
ref_mean = mean_values[0]

# --- Plot only selected temperatures (25, 45, 60) ---
selected_temps = {25, 31, 35, 40, 45, 50, 55, 60}

for (temp, time_ps, counts), mean_val in zip(curves, mean_values):
    if temp in selected_temps:
        normalized_counts = counts * (ref_mean / mean_val)
        fig.add_trace(go.Scatter(
            x=time_ps/1e9,
            y=normalized_counts,
            mode='lines',
            name=f"{temp}°C"
        ))

# --- Layout ---
fig.update_layout(
    template=fig_template,
    xaxis_title="Time (ms)",
    yaxis_title="Counts per bin (arb. unit)",
    legend_title="Temperature",
    legend=dict(font=dict(size=26)),
    width=900,
    height=600,
    margin=dict(l=110, r=50, t=50, b=80),
)

fig.show()
```

```python
import os
import numpy as np
import plotly.graph_objects as go
from scipy.stats import linregress

# --- Custom Plotly Template ---
fig_template = go.layout.Template()
fig_template.layout = {
    'template': 'simple_white+presentation',
    'autosize': False,
    'width': 800,
    'height': 600,
    'xaxis': {'ticks': 'inside', 'mirror': 'ticks', 'linewidth': 2.5, 'tickwidth': 2.5, 'ticklen': 4,
              'showline': True, 'showgrid': False, 'zerolinecolor': 'white',
              'titlefont': dict(size=34), 'tickfont': dict(size=30)},
    'yaxis': {'ticks': 'inside', 'mirror': 'ticks', 'linewidth': 2.5, 'tickwidth': 2.5, 'ticklen': 4,
              'showline': True, 'showgrid': False, 'zerolinecolor': 'white',
              'titlefont': dict(size=34), 'tickfont': dict(size=30)},
    'font': {'family': 'mathjax', 'size': 30}
}

# --- Folder and Files ---
folder_path = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\2nd Oct NaBiF4\temperature sensing"
temperatures = [25, 31, 35, 40, 45, 50, 55, 60]
use_2_suffix = {35, 45, 50, 55}

# --- Load data ---
curves = []
for temp in temperatures:
    fname = f"1064nm_3.84mW_20Hz_975nm_220mW_{temp}degree{'_2' if temp in use_2_suffix else ''}.txt"
    fpath = os.path.join(folder_path, fname)
    if not os.path.exists(fpath):
        print(f"⚠️ File not found: {fname}")
        continue

    data = np.loadtxt(fpath, skiprows=1, delimiter='\t')
    time_ps, counts = data[:, 0], data[:, 1]
    curves.append((temp, time_ps, counts))

# --- Plot raw curves (no normalization) ---
fig = go.Figure()
for temp, time_ps, counts in curves:
    fig.add_trace(go.Scatter(
        x=time_ps/1e9,
        y=counts,
        mode='lines',
        name=f"{temp}°C"
    ))

fig.update_layout(
    template=fig_template,
    xaxis_title="Time (ms)",
    yaxis_title="Counts per bin",
    legend_title="Temperature",
    legend=dict(font=dict(size=26)),
    width=900, height=600,
    margin=dict(l=110, r=50, t=50, b=80),
)
fig.show()

# --- Time groups (ms → ps) ---
groups_ps = [
    (21e9, 24e9),   # Group 1
    (46e9, 49e9),   # Group 2
    (71e9, 74e9),   # Group 3
    (96e9, 99e9),   # Group 4
]

# --- Histogram and Signal Analysis ---
signal_means, signal_stds = [], []

for temp, time_ps, counts in curves:
    # group extraction directly from raw counts
    group_vals = []
    for (t1, t2) in groups_ps:
        mask = (time_ps >= t1) & (time_ps <= t2)
        group_vals.append(counts[mask])

    g1, g2, g3, g4 = group_vals
    hist1_data = np.concatenate([g1, g3])  # 1st + 3rd
    hist2_data = np.concatenate([g2, g4])  # 2nd + 4th

    mean1, std1 = np.mean(hist1_data), np.std(hist1_data)
    mean2, std2 = np.mean(hist2_data), np.std(hist2_data)

    signal = mean1 / mean2
    signal_std = signal * np.sqrt((std1/mean1)**2 + (std2/mean2)**2)

    signal_means.append((temp, signal, signal_std))

    # --- Plot histograms per temperature ---
    hist_fig = go.Figure()
    hist_fig.add_trace(go.Histogram(x=hist1_data, name="Group 1+3", opacity=0.6))
    hist_fig.add_trace(go.Histogram(x=hist2_data, name="Group 2+4", opacity=0.6))
    hist_fig.update_layout(
        template=fig_template,
        barmode='overlay',
        title=f"Histograms at {temp}°C",
        xaxis_title="Counts per bin",
        yaxis_title="Frequency",
        legend=dict(font=dict(size=24))
    )
    hist_fig.show()

# --- Signal vs Temperature ---
temps = np.array([t for t, _, _ in signal_means])
signals = np.array([s for _, s, _ in signal_means])
sig_stds = np.array([e for _, _, e in signal_means])

# Linear Fit
slope, intercept, r_val, p_val, std_err = linregress(temps, signals)
fit_line = intercept + slope * temps

# --- Plot ---
fig_signal = go.Figure()
fig_signal.add_trace(go.Scatter(
    x=temps, y=signals, mode='markers',
    error_y=dict(type='data', array=sig_stds, visible=True),
    name="Data"
))
fig_signal.add_trace(go.Scatter(
    x=temps, y=fit_line, mode='lines',
    name=f"Fit: y = {slope:.3e}x + {intercept:.3f} (R²={r_val**2:.3f})",
    showlegend = False
))
fig_signal.update_layout(
    template=fig_template,
    xaxis_title="Temperature (°C)",
    yaxis_title="Signal (C₁/C₂)",
    # title="Signal vs Temperature (raw counts)",
    legend=dict(font=dict(size=24)),
    width=800,
    height=600,
    margin=dict(l=110, r=50, t=50, b=80),
)
fig_signal.show()
```

```python
import numpy as np

# --- Given parameters ---
integration_time_ms = 300
averages = 1
integration_time_s = integration_time_ms / 1000  # convert to seconds

# --- Calculate signal sensitivity ---
# slope (dS/dT) from the previous linear regression
dS_dT = slope
sigma_S = np.std(signals - fit_line)  # residual noise

# effective measurement rate
R = integration_time_s * averages

# sensitivity formula (°C per sqrt(Hz))
eta_T = sigma_S / (abs(dS_dT) * np.sqrt(R))

print(f"Slope (dS/dT): {dS_dT:.3e}")
print(f"Noise in signal (σS): {sigma_S:.3e}")
print(f"Effective measurement rate (R): {R:.3e}")
print(f"Temperature Sensitivity (ηT): {eta_T:.3e} °C·Hz⁻¹/²")
```
