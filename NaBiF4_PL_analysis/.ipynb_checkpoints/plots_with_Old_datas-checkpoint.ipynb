## Power variation of 1064nm laser

```python
# --- Folder containing the data files ---
data_folder = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\23 august\clean Particle\1064nm power variation\27th August"

# --- Calibration: current -> measured power (as you provided)
# currents (units are the numbers you gave) and their corresponding power strings
currents = np.array([1.25, 1.5, 2.0, 2.5, 3.0, 3.5], dtype=float)
power_strings = ["220uW", "900uW", "2.3mW", "4.2mW", "6.9mW", "10mW"]

# convert calibration power strings to watts (floats)
def power_str_to_watts(s):
    s = s.strip().lower()
    if s.endswith('uw'):
        return float(s.replace('uw','')) * 1e-6
    if s.endswith('mw'):
        return float(s.replace('mw','')) * 1e-3
    if s.endswith('w'):
        return float(s.replace('w',''))
    raise ValueError("Unknown power unit in calibration: " + s)

powers_w = np.array([power_str_to_watts(s) for s in power_strings], dtype=float)

# --- Initialize figure ---
fig = go.Figure()

# Add legend header as an invisible trace (to act like a header)
fig.add_trace(go.Scatter(
    x=[None], y=[None],
    mode="lines",
    line=dict(color="rgba(0,0,0,0)"),
    showlegend=True,
    name="<b>1064 nm Laser Power</b>"
))

# Store all datasets for normalization
datasets = []

for filename in os.listdir(data_folder):
    if filename.endswith(".txt"):
        filepath = os.path.join(data_folder, filename)

        # Extract the numeric token (the value you currently treat as "power" in the filename)
        # This will find numbers followed by "mW" or just numbers -- adapt if your filenames differ.
        # We'll use the first numeric token found that is reasonable (<= 10 for your currents).
        numeric_tokens = re.findall(r"(\d+\.?\d*)", filename)
        current_value = None
        for tok in numeric_tokens:
            val = float(tok)
            # heuristic: your currents lie in the range ~1.253.5; pick the first token in [0.5, 5]
            if 0.5 <= val <= 5.0:
                current_value = val
                break
        if current_value is None:
            # fallback: try to find pattern like 'X.XmW' from earlier method
            match = re.search(r"(\d+\.?\d*)mW", filename)
            if match:
                current_value = float(match.group(1))
            else:
                current_value = np.nan

        # Interpolate current -> power (watts). If current is outside calibration range, np.interp will
        # linear-extrapolate by default using end values (we keep that behavior).
        if not np.isnan(current_value):
            power_w = np.interp(current_value, currents, powers_w)
        else:
            power_w = np.nan

        # Format power label nicely: show as W if <1 mW else show mW
        if np.isfinite(power_w):
            if power_w < 1e-3:
                power_label = f"{power_w*1e6:.0f} W"    # integer W
            else:
                power_label = f"{power_w*1e3:.2f} mW"   # two decimals mW
        else:
            power_label = "unknown"

        # Read file with header (your files have header "Time (ps)"  "Counts per bin")
        df = pd.read_csv(filepath, sep="\t", header=0)
        df.rename(columns={"Time (ps)": "time_ps", "Counts per bin": "counts"}, inplace=True)

        # Ensure numeric and drop NaNs
        df["time_ps"] = pd.to_numeric(df["time_ps"], errors="coerce")
        df["counts"] = pd.to_numeric(df["counts"], errors="coerce")
        df = df.dropna()

        datasets.append((current_value, power_label, df))

# --- Reference mean (first dataset, 025 ms) ---
tmax_ps = 25e9  # 25 ms in ps

if len(datasets) == 0:
    raise RuntimeError("No datasets found in folder.")

ref_mean = datasets[0][2].loc[datasets[0][2]["time_ps"] <= tmax_ps, "counts"].mean()

# --- Optionally sort datasets by interpolated power (ascending) to get a nicer legend order ---
# sort key: numeric power in watts; if unknown, treat as large so they go to end
def power_w_from_label(lbl):
    if lbl == "unknown":
        return np.inf
    if "" in lbl:
        return float(lbl.split()[0]) * 1e-6
    if "mW" in lbl:
        return float(lbl.split()[0]) * 1e-3
    # default
    return np.inf

datasets_sorted = sorted(datasets, key=lambda x: power_w_from_label(x[1]))

# --- Add datasets with scaling and label their legend with interpolated power and scaling factor ---
for current_value, power_label, df in datasets_sorted:
    mean_val = df.loc[df["time_ps"] <= tmax_ps, "counts"].mean()
    scale_factor = ref_mean / mean_val if (mean_val is not None and mean_val != 0) else 1.0

    df_scaled = df.copy()
    df_scaled["counts"] *= scale_factor

    # show time axis in ms for plotting
    fig.add_trace(go.Scatter(
        x=df_scaled["time_ps"] / 1e9,   # ps -> ms
        y=df_scaled["counts"],
        mode="lines",
        name=f"{power_label} (scaled by {scale_factor:.2f})"
    ))

# --- Final layout ---
fig.update_layout(
    template=fig_template,   # assumes you have this defined already
    width=1200,
    height=600,
    margin=dict(l=120, r=50, t=50, b=80),
    xaxis_title="Time (ms)",
    yaxis_title="Photon counts (arb. unit)"
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from scipy.optimize import curve_fit

# --- Select the last dataset ---
current_value, power_label, df = datasets_sorted[-1]

# --- Convert time to ms ---
time_ms = df["time_ps"] / 1e9
counts = df["counts"]

# --- Select fitting window: 2449 ms ---
fit_mask = (time_ms >= 24) & (time_ms <= 49)
t_fit = time_ms[fit_mask]
y_fit = counts[fit_mask]

# --- Fixed lifetimes ---
tau1_fixed = 1.46   # ms
tau2_fixed = 6.89   # ms

# --- Define double exponential with fixed taus ---
def double_exp_fixed_tau(t, A1, A2, C):
    return A1 * np.exp(-t/tau1_fixed) + A2 * np.exp(-t/tau2_fixed) + C

# --- Initial guess for amplitudes and offset ---
p0 = [np.max(y_fit)/2, np.max(y_fit)/2, 0.0]

# --- Fit only for A1, A2, C ---
popt, pcov = curve_fit(double_exp_fixed_tau, t_fit, y_fit, p0=p0, maxfev=10000)
A1_fit, A2_fit, C_fit = popt

# --- Compute fitted curve within fitting range only ---
y_fitted = double_exp_fixed_tau(t_fit, *popt)

# --- Plot ---
fig = go.Figure()

# Original raw curve
fig.add_trace(go.Scatter(
    x=time_ms,
    y=counts,
    mode="lines",
    name=f"{power_label}",
    line=dict(width=2)
))

# Fitted double exponential (only within 2449 ms)
fig.add_trace(go.Scatter(
    x=t_fit,
    y=y_fitted,
    mode="lines",
    line=dict(dash="dash", color="firebrick", width=4),
    name=f"Double Exp Fit (1={tau1_fixed} ms, 2={tau2_fixed} ms)"
))

# --- Layout ---
fig.update_layout(
    template=fig_template,
    width=800,
    height=600,
    margin=dict(l=120, r=50, t=50, b=80),
    xaxis_title="Time (ms)",
    yaxis_title="Photon counts (arb. unit)",
    legend=dict(
        x=0.43, y=1.05,        # top center
        xanchor="center",
        yanchor="bottom",
        bgcolor="rgba(255,255,255,0.5)"
    )
)

fig.show()
```

```python
# --- Percentage change analysis (absolute values, no colorbar) ---

# Define which currents correspond to which target times (in ms)
target_times_ms_map = {
    49.2: [3.5, 3.25, 2.75, 2.0],
    47.0: [3.0, 3.25, 2.25,2.5]
}

# Build a lookup for quick access
current_to_target_time = {}
for t_ms, currents_list in target_times_ms_map.items():
    for c in currents_list:
        current_to_target_time[c] = t_ms

results = []

for current_value, power_label, df in datasets_sorted:
    if current_value not in current_to_target_time:
        continue  # skip datasets not in your selection

    # Compute average value between 025 ms
    avg_val = df.loc[df["time_ps"] <= 25e9, "counts"].mean()

    # Get target time in ps
    target_time_ms = current_to_target_time[current_value]
    target_time_ps = target_time_ms * 1e9

    # Find closest time index
    idx = (df["time_ps"] - target_time_ps).abs().idxmin()
    target_val = df.loc[idx, "counts"]

    # Percentage change (absolute value)
    perc_change = abs((target_val - avg_val) / avg_val * 100)

    # Extract numeric power in mW for plotting
    if "" in power_label:
        power_mW = float(power_label.split()[0]) * 1e-3
    elif "mW" in power_label:
        power_mW = float(power_label.split()[0])
    else:
        power_mW = np.nan

    results.append((power_mW, perc_change, power_label, current_value))

# --- Convert to DataFrame for convenience ---
df_results = pd.DataFrame(results, columns=["Power_mW", "PercChange", "PowerLabel", "Current"])

# --- Define power-law function ---
def power_law(x, a, b):
    return a * (x ** b)

# Remove NaNs
df_fit = df_results.dropna(subset=["Power_mW", "PercChange"])
xdata = df_fit["Power_mW"].values
ydata = df_fit["PercChange"].values

# --- Log-transform data ---
log_x = np.log(xdata)
log_y = np.log(ydata)

# --- Linear fit in log-log space ---
coeffs = np.polyfit(log_x, log_y, 1)  # slope = b, intercept = log(a)
b_fit = coeffs[0]
log_a_fit = coeffs[1]
a_fit = np.exp(log_a_fit)

# --- Generate smooth log-spaced x for plotting fitted curve ---
x_fit = np.logspace(np.log10(min(xdata)), np.log10(max(xdata)), 200)
y_fit = a_fit * x_fit**b_fit

# --- Create log-log figure ---
fig2 = go.Figure()

# Data points
fig2.add_trace(go.Scatter(
    x=xdata,
    y=ydata,
    mode="markers",
    marker=dict(size=10, color="firebrick", symbol="circle"),
    showlegend=False,
    text=[f"Current={c}, {pl}" for c, pl in zip(df_fit["Current"], df_fit["PowerLabel"])],
    hovertemplate="Power: %{x:.2f} mW<br>|%|: %{y:.2f}%<br>%{text}<extra></extra>"
))

# Fitted curve
fig2.add_trace(go.Scatter(
    x=x_fit,
    y=y_fit,
    mode="lines",
    line=dict(color="royalblue", width=3),
    name=f"fitted with y &#8733; x<sup>{b_fit:.2f}</sup>"
))

# --- Layout (log-log) ---
fig2.update_layout(
    template=fig_template,
    width=800,
    height=600,
    margin=dict(l=100, r=50, t=50, b=100),
    xaxis=dict(
        title="1064 nm Laser Power (mW)",
        type="log",
        autorange=True
    ),
    yaxis=dict(
        title="Absolute Percentage Change (%)",
        type="log",
        autorange=True
    ),
    legend=dict(x=0.02, y=0.98)
)

fig2.show()
```

## lifetimes with 975nm powers

```python
import numpy as np
import plotly.graph_objects as go
import os
import re
from scipy.optimize import curve_fit

# --- Path to folder ---
folder_path = r"C:\Users\Atanu\OneDrive\Desktop\research_work\paper\NaBiF4_PL_analysis\datas\23 august\clean Particle"

# --- Regex for files of interest ---
pattern = re.compile(r"975nm\s*(\d+)mW\.txt")

# --- Collect matching files ---
files_with_power = []
for fname in os.listdir(folder_path):
    match = pattern.match(fname)
    if match:
        power = int(match.group(1))  # numeric power
        files_with_power.append((power, fname))

# --- Sort by power ---
files_with_power.sort(key=lambda x: x[0])

# --- Define single exponential function ---
def exp_decay(t, A, tau, C):
    return A * np.exp(-t / tau) + C

# --- Initialize figure for decays ---
fig_decay = go.Figure()

# --- Store lifetimes and errors ---
powers = []
taus = []
sigma_taus = []

# --- Loop through sorted files ---
n_files = len(files_with_power)
for idx, (power, fname) in enumerate(files_with_power):
    file_path = os.path.join(folder_path, fname)

    # --- Load data with NumPy (skip header row) ---
    data = np.loadtxt(file_path, skiprows=1)
    time_ps = data[:, 0]   # Time in ps
    counts = data[:, 1]    # Counts per bin

    # --- Convert time to microseconds ---
    time_us = time_ps * 1e-6  

    # --- Normalize intensity ---
    counts_norm = counts / np.max(counts)

    # --- Fit exponential decay ---
    try:
        p0 = [1.0, np.median(time_us), 0.0]  # initial guess
        popt, pcov = curve_fit(exp_decay, time_us, counts_norm, p0=p0, maxfev=10000)
        A_fit, tau_fit, C_fit = popt
        sigma_tau = np.sqrt(np.abs(pcov[1,1])) if pcov is not None else 0.0

        # --- Store lifetime and its uncertainty ---
        powers.append(power)
        taus.append(tau_fit)
        sigma_taus.append(sigma_tau)

        # --- Plot only every 3rd file starting from last ---
        if (n_files - 1 - idx) % 3 == 0:
            # Original normalized decay
            fig_decay.add_trace(go.Scatter(
                x=time_us,
                y=counts_norm,
                mode="lines",
                name=f"{power} mW (={tau_fit:.2f} s)"
            ))

            # Fitted curve (dashed)
            fit_y = exp_decay(time_us, *popt)
            fig_decay.add_trace(go.Scatter(
                x=time_us,
                y=fit_y,
                mode="lines",
                line=dict(dash="dash"),
                name=f"Fit {power} mW",
                showlegend=False
            ))

    except RuntimeError:
        print(f" Fit failed for {fname}")
        sigma_taus.append(0.0)

# --- Layout for decay curves ---
fig_decay.update_layout(
    template=fig_template,
    width=800,
    height=600,
    xaxis_title="Time (s)",
    yaxis_title="Normalized Counts (a.u.)",
    margin=dict(l=100, r=50, t=50, b=100),
    legend=dict(
        x=0.45,
        y=0.95
    )
)

fig_decay.show()


# --- Lifetime vs Power plot with error bars (log y-axis) ---
fig_tau = go.Figure()

fig_tau.add_trace(go.Scatter(
    x=powers,
    y=taus,
    mode="markers",  # only markers, no connecting lines
    marker=dict(size=12, color="firebrick"),
    error_y=dict(
        type='data',
        array=sigma_taus,
        visible=True,
        thickness=2,
        width=6
    ),
    name="Lifetime "
))

# --- Layout ---
fig_tau.update_layout(
    template=fig_template,
    width=800,
    height=600,
    xaxis_title="975 nm Power (mW)",
    yaxis_title="Lifetime  (s)",
    yaxis_type="log",  # <-- logarithmic y-axis
    margin=dict(l=100, r=50, t=50, b=100)
)

fig_tau.show()
```
