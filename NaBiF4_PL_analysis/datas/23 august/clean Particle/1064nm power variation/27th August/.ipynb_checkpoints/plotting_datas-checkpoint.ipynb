```python
import os
import pandas as pd
import plotly.graph_objects as go

# Directory containing the data files
data_dir = r'D:\Data\2025\August\21 Aug Tm UCP spectra dual beam Agniva\Bi UCP Red emission\23 august\clean Particle\1064nm power variation\27th August'

# Get all .txt files in the folder
files = [f for f in os.listdir(data_dir) if f.endswith('.txt')]

# Create a Plotly figure
fig = go.Figure()

# Loop through all files and plot
for file in files:
    filepath = os.path.join(data_dir, file)
    
    # Read the data, skip header, tab/space delimited
    data = pd.read_csv(filepath, skiprows=1, sep=r'\s+', header=None)

    data.columns = ['Time_ps', 'Counts_per_bin']
    
    # Normalize the Counts_per_bin column
    data['Counts_per_bin'] = data['Counts_per_bin'] / data['Counts_per_bin'].max()
    
    # Convert time from picoseconds to microseconds for plotting
    data['Time_us'] = data['Time_ps'] / 1000.0

    # Add each trace to the figure
    fig.add_trace(go.Scatter(
        x=data['Time_us'],
        y=data['Counts_per_bin'],
        mode='lines',
        name=os.path.splitext(file)[0]  # Show file name without .txt
    ))

# Apply the given layout template
fig.update_layout(
    template='simple_white+presentation',
    autosize=False,
    width=1200,
    height=600,
    xaxis=dict(
        title='Delay  (s)',
        ticks='inside',
        mirror='ticks',
        linewidth=2.0,
        tickwidth=2.0,
        ticklen=6,
        showline=True,
        showgrid=False,
        zerolinecolor='white'
    ),
    yaxis=dict(
        title='lifetime with 1064nm (normalized)',
        ticks='inside',
        mirror='ticks',
        linewidth=2.0,
        tickwidth=2.0,
        ticklen=6,
        showline=True,
        showgrid=False,
        zerolinecolor='white'
    ),
    font=dict(
        family='mathjax',
        size=22,
    )
)

# Show the figure
fig.show()
```
